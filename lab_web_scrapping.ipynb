````python
# Initialize lists to hold the data
categories = []
titles = []
prices = []
availabilities = []
ratings = []
descriptions = []
upcs = []

# Loop through each book entry
# Attempt to safely extract the category
for book in soup.find_all('article', class_='product_pod'):
    # Title
    title = book.h3.a['title']
    titles.append(title)
    
    # Extract category from the URL structure
    href_parts = book.find('a').get('href').split('/')
    
    if len(href_parts) > 3:  # Ensure the list has enough parts
        category = href_parts[3]
    else:
        category = "Unknown"  # Fallback if structure is different
    
    categories.append(category)
    
    # Continue with other data extraction...
    price = book.find('p', class_='price_color').text
    prices.append(price)
    
    availability = book.find('p', class_='instock availability').text.strip()
    availabilities.append(availability)
    
    rating = book.p['class'][1]
    ratings.append(rating)
    
    description = "N/A"  # Placeholder for description
    descriptions.append(description)
    
    upc = "N/A"  # Placeholder for UPC
    upcs.append(upc)

````

````python
# Create a DataFrame from the lists
df = pd.DataFrame({
    'Category': categories,
    'Title': titles,
    'Price': prices,
    'Stock Availability': availabilities,
    'Star Rating': ratings,
    'Description': descriptions,
    'UPC': upcs
})
````

````python
# Clean the price column
df['Price'] = df['Price'].replace({'£': '', ',': ''}, regex=True).astype(float)
````

Start with the following dictionary:

Then, iterate over all the categories and all the books in a given category to collect any book information using the previous function. Fill the previous dictionary with the information about each book.

Show the first five rows of the previous final Pandas DataFrame.

````python
import pandas as pd
from tqdm import tqdm

# Initialize the dictionary to hold book data
books_dict = {
    "Title": [],
    "Price": [],
    "Availability": [],
    "Rating": [],
    "Description": [],
    "UPC": [],
    "Category": []
}

# Example categories and URLs (replace with actual URLs)
categories_urls = {
    "Travel": "https://books.toscrape.com/catalogue/category/books/travel_2/index.html",
    "Mystery": "https://books.toscrape.com/catalogue/category/books/mystery_3/index.html"
    # Add more categories and their URLs here
}

# Iterate over each category
for category, category_url in tqdm(categories_urls.items(), desc="Categories"):
    # Fetch the category page
    response = requests.get(category_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Get all book URLs in this category
    book_links = soup.find_all('h3')
    base_url = 'https://books.toscrape.com/catalogue/'
    book_urls = [base_url + link.find('a')['href'] for link in book_links]
    
    # Iterate over each book URL and get details
    for book_url in tqdm(book_urls, desc=f"Books in {category}"):
        details = get_book_details(book_url)
        
        # Append each detail to the corresponding list in books_dict
        books_dict["Title"].append(details["Title"])
        books_dict["Price"].append(details["Price"])
        books_dict["Availability"].append(details["Availability"])
        books_dict["Rating"].append(details["Rating"])
        books_dict["Description"].append(details["Description"])
        books_dict["UPC"].append(details["UPC"])
        books_dict["Category"].append(category)  # Add the category name

# Converting dictionary to a DataFrame
books_df = pd.DataFrame(books_dict)

# Display the first five rows of the DataFrame
print(books_df.head())

# Save the DataFrame to a CSV file
books_df.to_csv('books_data.csv', index=False)
````
